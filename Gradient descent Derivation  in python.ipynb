```python
w1=.15;
w2=.20;
w3=.25;
w4=.30;
w5=.40;
w6=.45;
w7=.50;
w8=.55;
i1=.05;
i2=.10;
b1=.35; 
b2=.60;
o1=.01;
o2=.99;
lamda=.50;
```

```python
neth1=i1*w1+i2*w2+b1*1;
outh1=1/(1+np.exp(-neth1));
neth2=i1*w3 + i2*w4+b1*1
outh2=1/(1+np.exp(-neth2))
```

```python
neto1=w5*outh1+w6*outh2+b2*1
outo1=1/(1+np.exp(-neto1))
neto2=w7*outh1+w8*outh2+b2
outo2=1/(1+np.exp(-neto2))
```

```python
"""Calculating eo1 with respect to outh1 we get"""
eo1outh1=-(o1-outo1)*outo1*(1-outo1)*w5*1.
"""Similarly calculating for e02 with respect to outh1 we get"""
eo2outh1=-(o2-outo2)*outo2*(1-outo2)*w7*1.
ew1=(w1)-lamda*(eo1outh1+eo2outh1)*outh1*(1-outh1)*i1
ew2=(w2)-lamda*(eo1outh1+eo2outh1)*outh1*(1-outh1)*i2
"""Calculating eo1 with respect to outh2 we get"""
eo1outh2=-(o1-outo1)*outo1*(1-outo1)*w6*1.
"""Similarly calculating for e02 with respect to outh2 we get"""
eo2outh2=-(o2-outo2)*outo2*(1-outo2)*w8*1.
ew3=(w3)-lamda*(eo1outh2+eo2outh2)*outh2*(1-outh2)*i1
ew4=(w4)-lamda*(eo1outh2+eo2outh2)*outh2*(1-outh2)*i2
print(ew1)
print(ew2)
print(ew3)
print(ew4)
```
